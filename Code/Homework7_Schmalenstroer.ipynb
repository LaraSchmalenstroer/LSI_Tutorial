### Write a function that outputs a dictionary with the names of the regions of the dog Cytochrome c gene and the corresponding sequence.

```python
def read_gene(path,name):
    with open(path+name, 'r') as file:
        data=file.read().replace('\n','').split('>')[1:]
    return data

def get_regions(gene):
    regions=[gene[i].split('protein_coding') for i in range(len(gene))]
    return regions

def dict_regions(gene):
    regions=get_regions(gene)
    dict_regions={regions[i][0]:regions[i][1] for i in range(len(regions))}
    return dict_regions

def name_exons(gene):
    regions=get_regions(gene)
    dict_names={'Exon'+str(i):regions[i][0] for i in range(len(regions)) if 'exon' in regions[i][0]}
    return dict_names

def get_exons(gene):
    regions=get_regions(gene)
    dict_exons={'Exon'+str(i):regions[i][1] for i in range(len(regions)) if 'exon' in regions[i][0]}
    return dict_exons

```

Write a function that calculates the possible number of permutations of a gene that is given as a list of its exons.

```python
def get_max_permutations(genes):
    """The function calculates the possible number of permutations of a given gene out of a list of exons. It returns \
    the number of permutations as an int value."""
    assert type(genes) is list or type(genes) is int, "The input must be a list of exons or a number of exons: %r" % genes
    import itertools
    permutations=[]
    if type(genes)==list:
        i=1
        while i > 0 and i <= len(genes):
            perm=itertools.permutations(genes, i)
            for p in perm:
                permutations.append(p)
            i+=1
    elif type(genes)==int:
        list_gene=[x for x in range(1,genes+1)]
        i=1
        while i > 0 and i <= len(list_gene):
            perm=itertools.permutations(list_gene, i)
            for p in perm:
                permutations.append(p)
            i+=1
    return len(permutations)
```
